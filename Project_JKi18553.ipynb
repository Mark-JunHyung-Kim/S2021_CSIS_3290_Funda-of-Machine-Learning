# 2. Data wrangling and transformation

### - Search error value & remove
- First Blood, Tower and Inhibitor means which team killed enemy hero, tower, or inhibitor first. 
- In this dataset, those columns (First blood, Tower, and Inhibitor) are relatively dependent between each team.
- This means that this values always occur at least in one of the teams (blue or red).
- So, it can not be 0(blue) & 0(red), and should be 1 & 0 or 0 & 1.

```python
# First blood, tower, and inhibitor values between blue and red team can not be 0 & 0
# So, search that values and try to remove

# set array of dependent feature names
blue_dep = ['blueFirstBlood', 'blueFirstTower', 'blueFirstInhibitor']
red_dep = ['redFirstBlood', 'redFirstTower', 'redFirstInhibitor']

# loop to search and remove error rows in the dataaset
for i in range(0, len(blue_dep)):
    cond_1 = (lol_df[blue_dep[i]] == 0)
    cond_2 = (lol_df[red_dep[i]] == 0)
    print('Both 0 of', blue_dep[i], '&', red_dep[i], ':', len(lol_df[cond_1 & cond_2]), 'rows are removed..')    
    lol_df.drop(lol_df[cond_1 & cond_2].index, inplace = True)
```

```python
# So now, those feature columns have opposite value each other (1 & 0 or 0 & 1) in all rows

# Check result
print(lol_df.shape)
lol_df[blue_dep + red_dep].head(5)
```

```python
# Now, those three dummied categorical variables became totally dependent.
# So, remove unneccesary column on the red team side (red FirstBlood, tower, and Inhibitor)

lol_df.drop(red_dep, axis = 1, inplace = True)
print(lol_df.columns)
lol_df.head(5)
```

```python
# Also, since the gameId and gameDuraton(duration) columns are unnecessary too, remove it.
# We already have original index column.
lol_df.drop(['gameId', 'gameDuraton'], axis = 1, inplace = True)

# After that, the Assist point can be interpreted as a lower value of kill & death, so also remove it.
lol_df.drop(['blueAssist', 'redAssist'], axis = 1, inplace = True)

# Each team's kill & death points are removed as they mean opposite values between the two teams.
# e.g. 'blueKills = redDeath' and 'blueDeath = redKills'
lol_df.drop(['redKills', 'redDeath'], axis = 1, inplace = True)
```

```python
# Then lastly, as we did before, blueWins(1) means lose of red team(redWins(0)), so cut the column off.
# and cut the blueWins column and re-concatenate it at the last part of feature dataframe for better readability.

# Delete 'redWins'
lol_df.drop('redWins', axis = 1, inplace=True)

# cut off 'blueWins'(y) and re-concatenate it at the last point of features dataframe.
response_y = lol_df['blueWins']
lol_df.drop('blueWins', axis = 1, inplace = True)
lol_df = pd.concat([lol_df, response_y], axis = 1)

# check dataset
lol_df.head(5)
```

```python
# Lastly, reset index of dataset for the deleted rows.
lol_df = lol_df.reset_index(drop = True)

# Check cleaned dataset.
print(len(lol_df))
lol_df.head(5)
```

### - Search & Remove outliers
 - For more accurate analysis and result, detect outliers and remove it.

```python
# create function to find outliers in given data

def find_outliers(data):
    
    # find Q1 and Q3 by using numpy - percentile function.
    # change result of numpy as normal value from array
    Q1 = np.percentile(data, [25])[0]    
    Q3 = np.percentile(data, [75])[0]    
    IQR = Q3 - Q1    
    
    # Set the MINIMUM AND MAXIMUM bounds.
    MIN_bound = Q1 - (IQR * 1.5)    
    MAX_bound = Q3 + (IQR * 1.5)
    
    print("Q1 : ", Q1)    
    print("Q3 : ", Q3)
    print("IQR : ", IQR)
    print("MINIMUM : ", MIN_bound)
    print("MAXIMUM : ", MAX_bound, "\n")
    
    # save indexes of outliers.
    index = data.loc[(data > MAX_bound) | (data < MIN_bound)].index
    
    # return the result.
    return index

def delete_outliers(data, feature):    
    
    print("Searching outliers...")
    outliers = find_outliers(data[feature])        
    data.drop(outliers, inplace = True)
    print("{} of outliers are removed in the {} column! \n\n".format(len(outliers), feature))
    
    return len(outliers)
```

```python
# create execution def to remove outliers with dataframe and target feature list array
def execute_remove_outlier(data, feature_list):
    
    total_removed_outliers = 0
    
    # print original dataset shape
    print("Size of dataframe before removing outliers: ", data.shape)
    
    # available maximum outlier rows to be removed is 10% (at least 90% of rows should be remained)
    print("Maximum ", round((data.shape[0] * 0.1),1), " rows can be removed! (Threshold: ", round((data.shape[0] * 0.9),1), " rows)\n" )
    
    for i in feature_list:
        total_removed_outliers += delete_outliers(data, i) 
    
    # print result of removing outlier and remained dataset
    print("Total ", total_removed_outliers, " of outliers are removed!")
    print("Size of dataframe after removing outliers: ", data.shape)
```

```python
# create target feature list to remove outliers
target_features = ['blueKills', 'blueDeath', 'blueTotalGold', 'blueAvgLevel']

# execute function
execute_remove_outlier(lol_df, target_features)
```

# 4. Machine learning pipeline implementations
 - Pipeline is powerful function to execute various analytic methods without any repeats.
 - In this section, we will create a base pipeline to implement scaling and classification.
 - Then, we will make second pipeline to find best hyperparameters for each classifier and accuracy scores to choice.

### - Make list for scalers and classifier for base pipeline
- We will use 3 different scalers (Standard, MinMax, and Robust) and 5 different classifiers (Logistic regression, KNN, SVM, Decision tree, and Naive Bayes)

```python
# create list for scalers and classifiers
scalers = [    
    ['Standard Scaler', StandardScaler()],
    ['MinMax Scaler', MinMaxScaler()],
    ['Robust Scaler', RobustScaler()]
]

classifiers = [
    ['Logistic Regression', LogisticRegression()],
    ['K-Nearest Neighborhood', KNeighborsClassifier()],
    ['Support Vector Machine', SVC()],
    ['Decision Tree Classifier', DecisionTreeClassifier()],
    ['Gaussian Naive Bayes', GaussianNB()]    
]
```

### - Create parameter lists to test
 - each parameter options will be used to the different classifier options

```python
# create empty parameter list for the default excution
option_params = []

# hyper parameters for LogisticRegression
params_logistic = [{   
    'clf__max_iter' : [50, 100, 200, 500], # maximum number of iteration    
    'clf__C' : [0.01, 0.1, 1, 5, 10], # strangth of regulation    
}]

# hyper parameters for KNeighborsClassifier 
param_knn = [{
    'clf__n_neighbors' : [1, 2, 3, 4],    # = K (normally, sqrt of number of features is good)
    'clf__p' : [1, 2] # p is type of method to calculate distance. p1 = Euclidean, p2 = Manhattan
}]

# hyper parameters for SVC with 'two different kernel types'
param_svm = [
    {'clf__kernel' : ['linear'], 'clf__C' : [10, 10, 100, 1000]},
    {'clf__kernel' : ['rbf'], 'clf__C' : [1.0, 10., 100., 1000.], 'clf__gamma': [0.01, 0.1, 1.0]}    
]

# hyper parameters for DecisionTreeClassifier
param_dt = [{
    'clf__max_depth' : [3, 4, 5] # maximum depth of branches
}]

# hyper parameters for Gaussian Naive Bayes
param_nb = [{
    'clf__var_smoothing' : [1e-9, 1e-8, 1e-7] # portion of the largest variance for calculation stability
}]
```

### - Create pipelines and running
 - By using loop and set_params statement, we can get different result from one execution.

```python
# create list for the scores
scaler_result = []
classifier_result = []
best_parameters = []
accuracy_scores = []

# create scaler list

# create classifier list. parameters for each classifier will be changed by using set_params of pipeline


# Loop, create pipeline, train and find out the model's accuracy
for sca in scalers:
    
    for clf in classifiers:
        
        class_pipe = Pipeline([
            ("scaler", sca[1]),
            ("clf", clf[1])
        ])
        
        # Cross Validation pipeline
        cv_pipe = Pipeline([
            ("gridCV", GridSearchCV(class_pipe, option_params, cv = 3, scoring = 'accuracy'))
        ])
        
        if clf[0] == 'Logistic Regression' :
            cv_pipe.set_params(gridCV__param_grid = params_logistic)
        elif clf[0] == 'K-Nearest Neighborhood' :
            cv_pipe.set_params(gridCV__param_grid = param_knn)
        elif clf[0] == 'Support Vector Machine' :
            cv_pipe.set_params(gridCV__param_grid = param_svm)
        elif clf[0] == 'Decision Tree Classifier' :
            cv_pipe.set_params(gridCV__param_grid = param_dt)
        elif clf[0] == 'Gaussian Naive Bayes' :
            cv_pipe.set_params(gridCV__param_grid = param_nb)
        
        # Fit pipe to the train dataset
        cv_pipe.fit(X_train, y_train)

        # store scaler and classifier names
        scaler = sca[0]
        clssifier = clf[0]
        
        # store best parameter and score of given scaler and classifier
        param = cv_pipe["gridCV"].best_params_
        score = cv_pipe["gridCV"].best_score_
        
        # save results into the lists
        classifier_result.append(clssifier)
        scaler_result.append(scaler)
        best_parameters.append(param)
        accuracy_scores.append(score)      
        

# create dataframe with result arrays
result_table = pd.DataFrame(zip(classifier_result, scaler_result, best_parameters, accuracy_scores), 
                            columns = ['Classifier', 'Scaler', 'Best HyperParameters', 'Best Accuracy Scores'])

# print (by execution order)
result_table
```

```python
# print (by best accuracy scores)
result_table.sort_values(by='Best Accuracy Scores', axis=0, ascending = False)
```

- According to the above result table, the best classifier for this dataset is **SVM (Support Vecter Vachine) with Robust Scaler.**
- Also, the best hyperparameter for classifier is **['C' : 100, 'gamma' : 0.01, 'kernel' : 'rbf']**

- So, next we will use this methods and hyperpameters to analyze results.

# 6. Out-of-sample prediction
 - Out of sample prediction means the predictions made by a model on data not used during the training of the model
 - In this out of sample prediction part, we will create new test data which is made by randonly created ranges to test.

### - Make definition to classify result as 'Win' or 'Lose'
 - If predicted result is 1, it means that blue team won the game
 - or, predicted result is 0, on the other hand, it means that blue team lose and red team won the game.

```python
# create result definition
def predict_result(y_pred):    
    if y_pred == 1:
        return "Win!!"    
    elif y_pred == 0:
        return "Lose.."        
```

### - Create random sample test data
 - All values are based on 1 or 0 (in case of categorical data), and Min and Max range (in case of numerical data)

```python
# create random values to predict win or lose

# set the size of numbers to test (randomly 1 - 20 times)
testsize = range(0, randint(1, 20))

# make array to store test sample values
X_new = []

# Create dataset for Out-of-sample prediction
for i in testsize:
    X_new.append([
        randint(0, 1), # blueFirstBlood (yes or no)
        randint(0, 1), # blueFirstTower (yes or no)
        randint(0, 1), # blueFirstInhibitor (yes or no)
        randint(min(features.blueKills), max(features.blueKills)), # blueKills (referenced by range of feature)
        randint(min(features.blueDeath), max(features.blueDeath)), # blueDeath (referenced by range of feature)
        randint(min(features.blueTotalGold), max(features.blueTotalGold)), # blueTotalGold (referenced by range of feature)
        round(uniform(min(features.blueAvgLevel), max(features.blueAvgLevel)), 1) # blueAvgLevel (referenced by range of feature)
    ])

# Check test samples
for i in X_new:
    print(i)
    
# create dataframe to store test samples
test_df = pd.DataFrame(X_new, columns = features.columns)
test_df
```

### - Test sample data and make prediction with best classification model
 - We already made pipeline which contains best model and hyperparameters.

```python
# create empty array to store result
clf_result = []

# the best pipeline name is best_pipe

# execute prediction with loop statement
for i in X_new:
    
    # predict result as 0 or 1
    y_pred_phi = best_pipe.predict([i])
    
    # convert result as Win or Lose string by using definition
    p_result = predict_result(y_pred_phi)
    
    # store result into result array
    clf_result.append(p_result)    
    
# create dataframe to store test sample results
response_df = pd.DataFrame(clf_result, columns = ['Win or Lose'])
response_df
```

### - Print table by sample data and prediction results

```python
# Attach the result dataframe to the sample dataframe
total_result = pd.concat([test_df, response_df], axis=1)

# print results
total_result
```
